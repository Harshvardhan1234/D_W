# fpgrowth

```python
from mlxtend.frequent_patterns import fpgrowth
import pandas as pd

# Sample dataset
dataset = [['milk', 'bread', 'eggs'],
           ['bread', 'butter'],
           ['milk', 'eggs', 'butter'],
           ['milk', 'bread', 'eggs', 'butter'],
           ['eggs', 'butter']]

# Convert dataset to pandas DataFrame
df = pd.DataFrame(dataset)
```

```python
df.columns = ["num1","num2","num3","num4"]
df
```

```python
dfx = pd.get_dummies(df["num3"])
```

```python
frequent_itemsets = fpgrowth(dfx, min_support=0.2, use_colnames=True)
frequent_itemsets
```

# apriori

```python
import pandas as pd
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules

# Create a sample dataset
dataset = {'Transaction_ID': ['T1', 'T2', 'T3', 'T4', 'T5'],
           'Items': [['milk', 'bread', 'eggs'],
                     ['bread', 'butter', 'cheese'],
                     ['milk', 'bread', 'butter'],
                     ['milk', 'bread', 'eggs', 'butter'],
                     ['milk', 'bread', 'eggs', 'cheese']]}

df = pd.DataFrame(dataset)

# Perform one-hot encoding on the 'Items' column
df_encoded = df.explode('Items')
df_encoded = pd.get_dummies(df_encoded['Items']).groupby(df['Transaction_ID']).sum()

# Apply Apriori algorithm to find frequent itemsets
frequent_itemsets = apriori(df_encoded, min_support=0.2, use_colnames=True)

# Extract association rules
association_rules = association_rules(frequent_itemsets, metric='confidence', min_threshold=0.5)


```
